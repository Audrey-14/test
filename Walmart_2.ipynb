```python
import pandas as pd

# Charger les ventes
df_sales = pd.read_csv("sales_train_validation.csv")

# "Melt" des colonnes de dates (d_1 à d_1913) en lignes
df_melted = pd.melt(
    df_sales,
    id_vars=['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'],
    var_name='d',
    value_name='sales'
)

# Exemple : affichage des 5 premières lignes
df_melted.head()
```

```python
df_melted.shape
```

```python
df_dates = pd.read_csv("calendar.csv")
df_dates.head(30)
```

```python
# Charger le calendrier
df_calendar = pd.read_csv("calendar.csv")

# Fusionner pour récupérer la vraie date
df_merged = df_melted.merge(df_calendar[['date','d',"wm_yr_wk"]], on='d', how='left')

# Conversion du champ 'date' en datetime
df_merged['date'] = pd.to_datetime(df_merged['date'])
```

```python
df_prices = pd.read_csv("sell_prices.csv")

df_prices.head(30)
```

```python
import pandas as pd

# Étape 1 : Calculer les ventes totales par item_id et par groupe
item_sales = (
    df_full.groupby(['cat_id', 'dept_id', 'store_id', 'item_id'])['sales']
    .sum()
    .reset_index(name='total_sales')
)

# Étape 2 : Calculer les ventes totales par groupe
group_sales = (
    item_sales.groupby(['cat_id', 'dept_id', 'store_id'])['total_sales']
    .sum()
    .reset_index(name='group_total_sales')
)

# Étape 3 : Ajouter la part de chaque groupe dans les ventes totales
item_sales = item_sales.merge(group_sales, on=['cat_id', 'dept_id', 'store_id'])
item_sales['group_sales_ratio'] = item_sales['group_total_sales'] / item_sales['total_sales'].sum()

# Étape 4 : Calculer combien de produits à échantillonner dans chaque groupe
item_sales = item_sales.merge(
    item_sales.groupby(['cat_id', 'dept_id', 'store_id'])['item_id']
    .nunique()
    .reset_index(name='n_items_in_group'),
    on=['cat_id', 'dept_id', 'store_id']
)

item_sales['group_sample_size'] = (item_sales['group_sales_ratio'] * 2000).round().astype(int)

# Étape 5 : Sélectionner les meilleurs vendeurs par groupe
sampled_item_ids = []
for (cat, dept, store), group_df in item_sales.groupby(['cat_id', 'dept_id', 'store_id']):
    group_df_sorted = group_df.sort_values(by='total_sales', ascending=False)
    sample_size = group_df['group_sample_size'].iloc[0]
    sampled = group_df_sorted.head(min(sample_size, len(group_df_sorted)))
    sampled_item_ids.extend(sampled['item_id'].tolist())

# Étape 6 : Filtrer df_full avec les item_id retenus
df_reduced = df_full[df_full["item_id"].isin(sampled_item_ids)].copy()
```
